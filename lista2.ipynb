```
def buildMatrix(tam):
    matrix = []
    for row in range(tam):
        row = list(map(int, input().split()))
        for num in range(len(row)):
            row[num] *= row[num]
        matrix.append(row)
    return matrix

def maior(matrix, col):
    m = 0
    for row in matrix:
        if len(str(row[col])) > m:
            m = len(str(row[col]))
    return m


def printMatrix(matrix):
    for row in range(len(matrix)):
        for col in range(len(matrix[row])):
            m = maior(matrix, col)
            valor = matrix[row][col]
            if (col != len(matrix)) and (col != 0):
                print(f' {valor:{m}d}', end='')
            else:
                print(f'{valor:{m}d}', end='')
        print()
    return None


n = int(input())
cont = 4
pos = 0
matrizes = []
for i in range(n):
    m = int(input())
    matrix = buildMatrix(m)
    matrizes.append(matrix)

while n != 0:
    print(f'Quadrado da matriz #{cont}:')
    printMatrix(matrizes[pos])
    pos += 1
    if n != 1:
        print()
    cont += 1
    n -= 1
```
